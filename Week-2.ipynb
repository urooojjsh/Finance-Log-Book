<html> <h3 style="color:blue;"> Grouping </h3> </html>

```python
# groupby() - the function groups the first column based on the second
```


.........................................

```python
Group_by_education = data.groupby(["sex", "education"])
Group_by_education.size()
```

```python
# As you see above, the string variables were not grouped correctly as they have some spaces.
```

.........................................

```python
data["sex"] = data["sex"].map(str.strip)
Group_by_education = data.groupby(["sex","education"])
Group_by_education.size()
```

```python
# The 'education' colomn also has data with spaces
```

.........................................

```python
data["education"] = data["education"].map(str.strip)
Group_by_education = data.groupby(["sex", "education"])
Group_by_education.size()
```

```python
# Now it's ok
```

.........................................

```python
Group_by_sex = data.groupby(['education', 'sex'])
print(type(Group_by_sex))
Group_by_education.size()
```

```python
# As you can see, Group_by_education is not a DataFrame, it has a special format - DataFrameGroupBy.

# To create a new DataFrame from DataFrameGroupBy, you need to apply the method unstack().
```

.........................................

```python
Group_by_sex_DataFrame = Group_by_sex.size().unstack()
Group_by_sex_DataFrame
```
